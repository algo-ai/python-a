代码膨胀了不少，如果更多的数呢，太笨重啦。有没有更好的办法呢？

想想平时我们怎么从一堆数中找出最大值的

我们可以先假定第一个数为当前的最大值，再依次和余下的每一个数进行比较，

如果比我们记录的最大值还大，那就把最大值改为该数，这样挨个比较完了，

我们就记录到它们的最大值啦。狗熊掰棒子，是不是就这么干的？

哈哈，这就是计算机常用的遍历方法

```python
# 接收用户输入的6个数，输出其中的最大值
a, b, c, d, e, f = eval(','.join(input().split()))
maxnum = a
if b>maxnum:
    maxnum = b
if c>maxnum:
    maxnum = c
if d>maxnum:
    maxnum = d
if e>maxnum:
    maxnum = e
if f>maxnum:
    maxnum = f
print(maxnum)
```

代码是不是简洁很多？

逐个逐个地把队列里的每个数据进行处理，不重复、不遗漏，最后就可以找到结果。

这就是遍历，也是常用的一种方法，要学会应用哦

我们可以看到，对于每一个数据，我们要做的处理是一样的，

但由于变量名各不相同，我们只能一遍又一遍地重复同样的操作

嗯，接下来我们就用列表和循环来找多个数中的最大值

```python
# 输出用户输入的若干整数中的最大值

a = input().split()      # 将用户输入的一行内容（字符串）默认按空格进行分割，放到列表中（字符串）
maxnum = int(a[0])       # 把列表第一个元素（字符串）用int()函数转为整数
for i in range(len(a)): # 构造列表编号的序列进行循环遍历
    temp = int(a[i])     # 把遍历到的元素（字符串）转为整数
    if temp>maxnum:     # 判断当前元素是否大于记录的最大值
        maxnum = temp   # 如是则将当前元素记录为最大值
print(maxnum)           # 结束循环遍历后，输出最大值
```
