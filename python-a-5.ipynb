6. 我们再来看看第二题P001 A+B Problem，这其实也是演示题，让大家熟悉输入输出格式的要求。千万要注意，网络答题时，要严格按照题目规定的输入输出格式，不需要输入提示的一定不写提示，输出内容也是，题目叫你输出一个数那就输出一个数，不要输出多于内容。


这道题其实已经列出了各种编程语言的示例程序。我们一起来看看python如何读取用空格隔开的两个数

```python
# 我们知道可以用input()函数读取用户输入的一行内容，是字符串的形式
s = input()
print(s)
print(type(s))  # type函数查看类型
```

可以看到，s的类型是str（字符串）

```python
# 读取输入的时候，可以用eval函数直接将数字字符串内容转成数值类型
a = eval(input())
print(a)
print(type(a))
```

可以看到，a的类型是int（整数）

```python
# eval函数还可以读取用逗号隔开的多个数字字符串
a, b = eval(input())
print(a)
print(b)
```

但是要注意，用户输入的必须是数字内容才能用eval来处理，否则会出错

```python
a = eval(input())
print(a)
```

输入123abc，无法转为数值，出错了

eval处理的多个数之间只能是逗号隔开。如果是空格或其他符号，会出错。

```python
a, b = eval(input())
print(a)
print(b)
```

那么，怎么获取空格隔开的两个数呢？

第一步，看示例程序，用split()方法可以把空格隔开的数分割开

```python
input().split()
```

第二步，我们再用逗号把它们连起来（join()方法）

```python
','.join(input().split())
```

第三步，我们就可以像原来一样用eval()来获取数值了

```python
a, b = eval(','.join(input().split()))
print(a+b)
```

这就是第二题P001 A+B Problem的一种解决方案

接下来我们再看第三题P1421 小玉买文具


读取用户输入的a元b角，这和上一题一样的呀。那得到a、b两个数之后，我们把具体的钱数算出来，再除以每支笔的价格，不就得到答案啦

注意要使用整数来运算，除法也应该用取商运算，我们总不能买0.5支笔吧

所以应该把a、b换算为角，也就是a*10+b角，用笔的单价19角取商，完美

```python
a, b = eval(','.join(input().split()))
print((a*10+b)//19)
```
